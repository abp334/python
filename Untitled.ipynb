```python
#Operator overriding
# class Demo:
#     def add(self,n1,n2):
#         print(n1+n2)
#     def add(self,n1,n2,n3):
#         print(n1+n2+n3)
# d=Demo()
# d.add(15,20,30)

#Operator overloading
class Demo:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def __str__(self):
        return "X "+str(self.x)+" Y "+str(self.y)  
    def __add__(self,other):
        x=self.x+other.x
        y=self.y+other.y
        return Demo(x,y)
    def __sub__(self,other):
        x=self.x-other.x
        y=self.y-other.y
        return Demo(x,y)
    def __mul__(self,other):
        x=self.x*other.x
        y=self.y*other.y
        return Demo(x,y)
    def __truediv__(self,other):
        x=self.x/other.x
        y=self.y/other.y
        return Demo(x,y)
    def __floordiv__(self,other):
        x=self.x//other.x
        y=self.y//other.y
        return Demo(x,y)
    def __mod__(self,other):
        x=self.x%other.x
        y=self.y%other.y
        return Demo(x,y)
    def __pow__(self,other):
        x=self.x**other.x
        y=self.y**other.y
        return Demo(x,y)
    def __lt__(self,other):
        x=self.x<other.x
        y=self.y<other.y
        return Demo(x,y)
    def __gt__(self,other):
        x=self.x>other.x
        y=self.y>other.y
        return Demo(x,y)
    def __ge__(self,other):
        x=self.x>=other.x
        y=self.y>=other.y
        return Demo(x,y)
    def __le__(self,other):
        x=self.x<=other.x
        y=self.y<=other.y
        return Demo(x,y)
    def __ne__(self,other):
        x=self.x!=other.x
        y=self.y!=other.y
        return Demo(x,y)
    def __eq__(self,other):
        x=self.x==other.x
        y=self.y==other.y
        return Demo(x,y)
d=Demo(6,31)
d1=Demo(3,4)
print(d+d1)
print(d-d1)
print(d*d1)
print(d/d1)
print(d//d1)
print(d%d1)
print(d**d1)
print(d>d1)
print(d<d1)
print(d>=d1)
print(d<=d1)
print(d!=d1)
# sub subtract
# mul multiply
# truediv  true division
# floordiv  floor division
# mod
# pow
# lt for less than
# gt for greater than
# ge for greater than and equals to
# le for less than and equals to
# ne not equals to
# eq for double equals to
```

```python
#WAPP That has class Point and location as Child class of point find reflection of point about x axis
#reflection method define in location class
class Point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
class Location(Point):
    def reflection(self):
        return self.x ,0-self.y
l=Location(2,-2)
print(l.reflection())
```
